```python
# Import Standard Libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Import Specific SciKit Modules
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split as tts
```

```python
# Ordinary Least Squares
# Well, this is a form of linear regression. A simple one
# We want to minimise the sum of the residuals (Squared)
```

```python
dataset = pd.read_csv("./course_assets/Part 2 - Regression/Section 4 - Simple Linear Regression/Python/Salary_Data.csv")
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:, -1].values

# Split the data into test and training
x_train, x_test, y_train, y_test = tts(x,y,test_size = 0.2, random_state=0)
```

```python
linear_regressor = LinearRegression()
linear_regressor.fit(x_train, y_train) # This trains the model. Very Fun
# We have now fitted the model.
```

```python
y_pred = linear_regressor.predict(x_test) # Run this model on a certain range of data
```

```python
# Plot this

plt.scatter(x_train, y_train, color='red')
plt.plot(x_train, linear_regressor.predict(x_train), color='blue')
plt.title("Salary vs Experience (Training Set)")
plt.xlabel("Years of Experience ")
plt.ylabel("Salery")
plt.show()
```

```python
plt.scatter(x_test, y_test, color='red')
# The Regression Line is FROM THE TRAINING SET. Turns out it would not do anything, because we FIT on the test
plt.plot(x_train, linear_regressor.predict(x_train), color='blue')
plt.title("Salary vs Experience (Test Set)")
plt.xlabel("Years of Experience ")
plt.ylabel("Salery")
plt.show()
```
